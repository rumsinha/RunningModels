```python
from spacy.en import English
```

```python
parser = English()
```

```python
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
```

```python
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

```python
nltk.download('stopwords')
en_stop = set(nltk.corpus.stopwords.words('english'))

```

#### Preparing the text for topic modelling

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

#### open up data, read line by line, for each line, prepare text for LDA then add to a list

```python
import random
text_data = []
for line in df['text']:
        tokens = prepare_text_for_lda(line)
        if random.random() > .99:
            print(tokens)
            text_data.append(tokens)
```

#### LDA with Gensim

#### First creating a dictionary from the data then convert to bag of words corpus and save the dictionary and corpus
'

```python
from gensim import corpora
dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
```

```python
import pickle
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

```python
import gensim
NUM_TOPICS = 5
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word = dictionary, passes=15)
ldamodel.save('model5.gensim')
```

```python
topics = ldamodel.print_topics(num_words=100)
for topic in topics:
    print(topics)
```
